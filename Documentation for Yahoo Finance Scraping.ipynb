# Stock Price Scraping Project Documentation

## Overview

This project is a Python script that scrapes historical stock prices from Yahoo Finance, allowing users to retrieve and visualize the past 1 month data for a specified stock symbol. The script utilizes web scraping techniques to extract data from the Yahoo Finance website.

## Prerequisites

- Python 3.x
- Required libraries: `requests`, `beautifulsoup4`, `pandas`, `matplotlib`

Install the required libraries using the following command:

```bash
pip install requests beautifulsoup4 pandas matplotlib

## Installation

1. **Clone the repository:**

    ```bash
    git clone https://github.com/your-username/stock-price-scraper.git
    ```

2. **Navigate to the project directory:**

    ```bash
    cd stock-price-scraper
    ```

3. **Run the script:**

    ```bash
    python stock_scraper.py
    ```

## Usage

1. Run the script.
2. Enter the stock symbol when prompted.
3. View the scraped data and the plot of the past 1 month stock prices.

## Code Structure

The main script file is `stock_scraper.py`, and it uses the following functions:

- `get_stock_data(stock_symbol)`: Fetches historical stock prices using web scraping.
- `plot_stock_data(df, stock_symbol)`: Plots the closing prices for the past 1 month.

## Data Sources

The project relies on Yahoo Finance as the data source for historical stock prices. Ensure that the specified stock symbol is valid and listed on Yahoo Finance.

## Troubleshooting

If you encounter issues fetching data or running the script, consider the following:

- Ensure a stable internet connection.
- Check if the specified stock symbol is valid and listed on Yahoo Finance.
- Verify that the required libraries are installed.

## Contributions

Contributions to the project are welcome. If you find any bugs, have feature requests, or want to contribute improvements, please open an issue or submit a pull request.

## License

This project is licensed under the MIT License.

## Acknowledgments

Special thanks to the developers of the `requests`, `beautifulsoup4`, `pandas`, and `matplotlib` libraries.
